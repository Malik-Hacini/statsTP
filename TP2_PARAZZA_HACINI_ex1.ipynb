#### a)

```r

n<-100
N<- 10000
p<-0.2

```

#### b)

```r

x<- rbinom(n,1,p)
empirical_means <- cumsum(x) / 1:n
```

#### c)

```r
options(repr.plot.width=12, repr.plot.height=8)
plot(1:n,empirical_means ,type = "l",
xlab = "Sample Size (k)" ,
ylab = "empirical means",
main = "Convergence of the Empirical Mean: SLLN")
abline(h = p, col = "red")

legend("topright", legend = c("Empirical Mean Curve","Theoretical Mean"),col = c("black", "red"), lwd = 2)
```

This plot is showing the Strong Law of Large Numbers. the expected value of the Bernouilli distribution is E(X)=p so the SLLN predict that the squence of empirical means $X_k$ will converge almost surrely to p.
This is exactly what we can observe graphically on the plot.

#### d)

Each time we run the script an other random sample is generated. Thus we still have the same convergence that's demonstrate the consistency of the SLLN even if the random sample generated changes the SLLN guarantees that the sample mean converge almost surrely to p.


#### e)

```r
matrice_echantillons <- matrix(rbinom(N * n, 1, p), nrow = N, ncol = n)
```

#### f)


```r
moyennes_Xn <- rowMeans(matrice_echantillons)
```

#### g)

```r
sigma_theorique <- sqrt(p * (1 - p) / n)
hist(moyennes_Xn,probability = TRUE,breaks = 30,
xlab = "Sample Mean" ,
ylab = "Density",
main = "Sampling Distribution of the Mean vs. the Normal (CLT)")
curve(dnorm(x, mean = p, sd = sigma_theorique), add = TRUE)

legend("topright", legend = c( "Normal Distribution"), lwd = c(2))
```

This plot is showing the Central Limit Theorem,the distribution of a normalized version of the sample mean converges to a standard normal distribution.

- for $n=100$ the approximation is precise. The factor n is the number of sample so the greater it is the better the approximation will be. 
    (for n around 10 we have a pretty bad approximation)
- N is the number of simulation.If we don't take enought simulation the data will not be able to have a good estimation of the distribution, for $N=100$ it's not representativ at all, at $N=1000$ it's a good approximation and starting from there we would not need more data to have a good estimator
- p is the center of symetrie of the curve


#### h)

```r
plot(ecdf(moyennes_Xn), main = "ECDF vs. Theoretical Normal CDF", xlab="Sample Mean",ylab= "Cumulative Probability")
curve(pnorm(x, mean = p, sd = sigma_theorique), 
      col = "red",add = TRUE)

legend("right", legend = c("ECDF", "Theoretical Normal CDF"),col = c("black", "red"), lwd = c(2,2)) 
```

#### i)


```r
normale <- rnorm(N,mean = p,sd =  sigma_theorique)
qqplot(normale ,moyennes_Xn,main = "Normal QQPlot of Sample Means" ,xlab = "Theoretical Quantiles",ylab = "Empirical Quantiles",pch = 20)
abline(a = 0, b = 1, col = "red", lwd = 2)
legend("topleft", legend = c("x=y"),col = c("red"),lwd = c(2))
```

We can see that the empirical quantiles align with the theoretical quantiles of the normal distribution showing that the distribution of the sample is approximated correctly by the gaussian distribution. This represent correctly the central limit theorem.


**Exercice 1 . 2**

```r
mu <- p
sigma <- sqrt(p * (1 - p))
x <- rnorm(n, mean = mu, sd = sigma)
empirical_means <- cumsum(x) / 1:n

plot(1:n,empirical_means ,type = "l",
xlab = "Sample Size (k)" ,
ylab = "empirical means",
main = "Convergence of the Empirical Mean: SLLN")
abline(h = p, col = "red")
legend("topright", legend = c("Empirical Mean Curve","Theoretical Mean"),col = c("black", "red"), lwd = 2)
```

```r

matrice_echantillons <- matrix(rnorm(N * n,mu,sigma), nrow = N, ncol = n)
moyennes_Xn <- rowMeans(matrice_echantillons)
sigma_theorique <- sqrt(p * (1 - p) / n)
hist(moyennes_Xn,probability = TRUE,breaks = 30,
xlab = "Sample Mean" ,
ylab = "Density",
main = "Sampling Distribution of the Mean vs. the Normal (CLT)")
curve(dnorm(x, mean = p, sd = sigma/sqrt(n)), add = TRUE)
legend("topright", legend = c( "Normal Distribution"), lwd = c(2))
```

```r
plot(ecdf(moyennes_Xn), main = "ECDF vs. Theoretical Normal CDF", xlab="Sample Mean",ylab= "Cumulative Probability")
curve(pnorm(x, mean = p, sd = sigma_theorique), 
      col = "red",add = TRUE)
legend("right", legend = c("ECDF", "Theoretical Normal CDF"),col = c("black", "red"), lwd = c(2,2)) 
```
